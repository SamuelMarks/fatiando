```python
"""                                                                             
GravMag: 3D inversion in spherical coordinates using harvester and tesseroids   
"""
from fatiando import gravmag, gridder, logger, mesher, utils
from fatiando.vis import mpl, myv
import cPickle as pickle
```

```python
log = logger.get()
```

```python
props = {'density':1000}
model = [mesher.Tesseroid(-58, -52, -22, -18, -100000, -1600000, props)]
```

```python
myv.figure(zdown=False)
myv.tesseroids(model, 'density')
myv.earth(opacity=0.4)
myv.continents()
myv.show()
```

```python
bounds = [-70, -40, -40, 0, -30000, -3000000]
shape = (100, 100)
lons, lats, heights = gridder.regular(bounds[:4], shape, z=250000)
pot = utils.contaminate(gravmag.tesseroid.potential(lons, lats, heights, model), 0.01, percent=True)
```

```python
mpl.figure(figsize=(10, 8))
bm = mpl.basemap(bounds[:4], 'ortho')
bm.drawcoastlines()
bm.drawmapboundary()
mpl.contourf(lons, lats, pot, shape, 30, basemap=bm)
cb = mpl.colorbar()
cb.set_label('SI')
```

```python
mesh = mesher.TesseroidMesh(bounds, (20, 50, 50))
```

```python
seeds = gravmag.harvester.sow([[-55, -20, -700000, props]], mesh)
```

```python
myv.figure(zdown=False)
myv.tesseroids(model, style='wireframe')
myv.tesseroids([mesh[s.i] for s in seeds])
myv.earth(opacity=0.4)
myv.continents()
myv.show()
```

```python
data = [gravmag.harvester.Potential(lons, lats, heights, pot, meshtype='tesseroid')]
```

```python
estimate, predicted = gravmag.harvester.harvest(data, seeds, mesh, compactness=100, threshold=0.001)
```

```python
mesh.addprop('density', estimate['density'])
```

```python
mpl.figure(figsize=(20, 8))
mpl.subplot(1, 2, 1)
bm = mpl.basemap(bounds[:4], 'ortho')
bm.drawcoastlines()
bm.drawmapboundary()
levels = mpl.contour(lons, lats, pot, shape, 10, basemap=bm, color='k')
mpl.contour(lons, lats, predicted[0], shape, levels, basemap=bm, color='r')

mpl.subplot(1, 2, 2)
mpl.hist(pot - predicted[0], bins=10)
```

```python
result = mesher.vremove(0, 'density', mesh)
```

```python
scene = myv.figure(zdown=False)
myv.tesseroids(model, 'density', style='wireframe')
myv.tesseroids([mesh[s.i] for s in seeds], 'density')
myv.tesseroids(result, 'density')
myv.earth(opacity=0.4)
myv.continents()
scene.scene.camera.position = [40571265.978711426, 7228019.1836120738, 91124.133411396644]
scene.scene.camera.focal_point = [-226904.33919478679, -5052729.9730009167, -1617401.9518302111]
scene.scene.camera.view_angle = 5.033164800000001
scene.scene.camera.view_up = [0.27853449438228844, -0.86819033637018272, -0.41068732056472335]
scene.scene.camera.clipping_range = [24201716.789816849, 62026373.704526894]
scene.scene.camera.compute_view_plane_normal()
scene.scene.render()
myv.show()
```
